## By Random Forest Regressor

```python
from sklearn.inspection import permutation_importance
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df2=df2.drop(columns=['IsHoliday','year'],axis=1)
```

```python
features=df2.drop(columns=['Weekly_Sales'])
target=df2['Weekly_Sales']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df2.drop(columns=['Weekly_Sales'])
y = df2['Weekly_Sales']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
```

```python
rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=18,
    min_samples_leaf=15,
    max_features='sqrt',
    n_jobs=-1,
    random_state=42
)
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

```python
y_pred = rf.predict(X_test)
```

```python
from math import sqrt
```

```python
print("mean squared error is : ",round(mean_squared_error(y_test,y_pred),2))
print("RMSE: ",sqrt(mean_squared_error(y_test, y_pred)))
print("r2 score is:",r2_score(y_test,y_pred))
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_test.values[:100], label="Actual")
plt.plot(y_pred[:100], label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Sales (Sample)")
plt.show()
```

## By Desicion Tree Regressor

```python
param_grid = {
    'max_depth': [5, 10, 15, None],
    'min_samples_leaf': [1, 5, 10],
    'min_samples_split': [2, 10, 20]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(
    DecisionTreeRegressor(random_state=42),
    param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1
)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.score(X_train,y_train)
```

```python
grid.score(X_test,y_test)
```

```python
y_pred1=grid.predict(X_test)
```

```python
print("R2 score  :",r2_score(y_test, y_pred1))
print("MSE score  :",mean_squared_error(y_test, y_pred1))
print("RMSE: ",sqrt(mean_squared_error(y_test, y_pred1)))
```

```python
plt.figure(figsize=(12,5))
plt.plot(y_test.values[:100], label="Actual")
plt.plot(y_pred1[:100], label="Predicted")
plt.legend()
plt.title("Actual vs Predicted Sales (Sample)")
plt.show()
```
